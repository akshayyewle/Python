# **Data Import**

```python
df = pd.read_csv('world_population.csv')

print(df.shape)
df.head()
```

## **Change Display Settings To Show All Rows**

```python
# Change Display Setting
pd.set_option('display.max.rows',20) # Max 10 Rows

# Display The Dataframe
df
```

# **Filtering By Numeric Value**

```python
df[df['Rank']<=10]
```

# **Isin Function Filtering**

```python
# Specify Countries For Selection
Selected_Countries = ['Bangladesh','India','China']

# Filter Dataframe
df[df['Country'].isin(Selected_Countries)]
```

# **Contains Function Filtering**

```python
# Filter Dataframe
df[df['Country'].str.contains('United')]
```

# **Filter Function**

* For Filtering By Row & Column Indices

```python
# Filter By Column Names
df.filter(['Country','Rank','Continent','Capital'],axis=1)
```

```python
# Filter By Row Values
df.filter(items=['Zambia','Zimbabwe'],axis=0)  # Troubleshoot
```

```python
# Copy The Dataframe & Use Country Column As Row Index
df_02 = df
df_02.set_index('Country',inplace=True)
df_02
```

```python
# Filter By Row Index
df_02.filter(items=['Zambia','Zimbabwe'],axis=0) 
```

# **loc & iloc Filtering**

```python
# Filtering By Row Index
df_02.loc['United States']
```

```python
# Filtering By Row Index Number
df_02.iloc[3]
```

# **Sort Values Function**

```python
# Sort Values By Single Column
df[df['Rank']<=10].sort_values(by='Rank')
```

```python
# Sort Values By Multiple Columns
df[df['Rank']<=10].sort_values(by=['Rank','Continent'])
```
