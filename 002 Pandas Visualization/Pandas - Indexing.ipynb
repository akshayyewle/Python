# **Data Import**

```python
df = pd.read_csv('world_population.csv')

print(df.shape)
df.head()
```

# **Select Index Col While Importing Data**

```python
df = pd.read_csv('world_population.csv',index_col='Country')

print(df.shape)
df.head()
```

# **Reset Index**

```python
df.reset_index(inplace=True)
df
```

# **Set Index Function**

```python
df.set_index('Country',inplace=True)
df
```

# **Loc Filtering**

```python
# Filtering With Index Labels
df.loc['Albania']
```

```python
# Filtering With Index Position
df.iloc[1]
```

# **Multi Indexing**

```python
df.reset_index(inplace=True)
df
```

```python
df.set_index(keys=['Continent','Country'],inplace=True)
df
```

# **Sort By Index**

```python
df.sort_index(ascending=True)
```

```python
# Multi Index Sorting
df.sort_index(ascending=[True,False],inplace=True)
df
```

# **Multi-Index Filtering**

```python
# Single Index Filtering
df.loc['Africa']
```

```python
# Multi Index Filtering
df.loc['Africa','Angola']
```
