## **Strip Plot**

* See distribution of data without the additional quartile details as shown in box plot

```python
sns.stripplot(data=tips,x='day',y='tip',hue='sex',palette='YlGnBu')
```

```python
# Enable Category Dodging
sns.stripplot(data=tips,x='day',y='tip',hue='sex',palette='YlGnBu',dodge=True)

# Separate Strip Plot For Each Hue Color
```

## **Joint Plot**
* https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sns.jointplot(data=tips,x='tip',y='total_bill')
```

```python
# Show Regression Line
sns.jointplot(data=tips,x='tip',y='total_bill',kind='reg')
```

```python
# Show KDE Lines
sns.jointplot(data=tips,x='tip',y='total_bill',kind='kde')
```

```python
# Show KDE Lines With Shading
sns.jointplot(data=tips,x='tip',y='total_bill',kind='kde',fill=True)

# Note: fill argument was earlier called shade
```

```python
# Show KDE Lines With Shading With Colour Scheme
sns.jointplot(data=tips,x='tip',y='total_bill',kind='kde',fill=True,cmap='RdYlGn')

# Note: Use cmap argument & not pallete
```

```python
# Show Hex Line
sns.jointplot(data=tips,x='tip',y='total_bill',kind='hex')
```

## **Penguin Dataset Example**

```python
penguins.head()
```

```python
sns.clustermap(data=penguins.select_dtypes('number').dropna())

# Note: Scale Of Values in "body_mass_g" column is different from values in other columns
# Hence, we need to standardize the values across all the columns
```

```python
sns.clustermap(data=penguins.select_dtypes('number').dropna(),standard_scale=True,metric='euclidean',figsize=(5,5))
```

```python
# Use Manhatten Distance As Clustering Metric
sns.clustermap(data=penguins.select_dtypes('number').dropna(),standard_scale=True,metric='cityblock',figsize=(5,5))
```

```python
# Use Correlation Distance As Clustering Metric
sns.clustermap(data=penguins.select_dtypes('number').dropna(),standard_scale=True,metric='correlation',figsize=(5,5))
```

## **Use Row Colour Option**

```python
penguins.head()
```

```python
# Unique Values In Species Column
penguins['species'].unique()
```

```python
# Use Row Colour Option
species_color = penguins.species.map({'Adelie':'blue','Chinstrap':'red','Gentoo':'green'})

print(type(species_color))
print(len(species_color))
```

```python
sns.clustermap(data=penguins.select_dtypes('number').dropna(),standard_scale=True,
               figsize=(7,7),
               row_colors=species_color)
```

### **Custom Palette To Highlight Specific Categories**

```python
tips.head()
```

```python
tips['day'].unique()
```

```python
Color_Palette_01 = {x: 'gray' for x in tips['day'].unique()}
print(Color_Palette_01)

# Add Highlighting Color
Color_Palette_01['Sat'] = 'crimson'
print(Color_Palette_01)
```

```python
# Scatterplot With Custom Palette
sns.scatterplot(tips,x='tip',y='total_bill',hue='day',palette=Color_Palette_01)
```

```python
# Displot With Default Palette
sns.displot(tips,x='tip',hue='day',col='day')
```

```python
# Displot With Custom Higlighting Palette
sns.displot(tips,x='tip',hue='day',col='day',palette=Color_Palette_01)
```
