# **Setup**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

# **Line Plot**

```python
# Define 1D Vectors
x = [1,2,3]
y = [2,4,6]
```

## **Plotting & Display Function**

```python
# Show Plot With Matplotlib Output Details
plt.plot(x,y)
```

```python
# Show Plot Without Matplotlib Output Details
plt.plot(x,y)
plt.show()
```

## **Add Plot & Axes Title**

```python
# Create Plot
plt.plot(x,y)

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Display Plot
plt.show()
```

## **Add Axes Ticks**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot
plt.plot(x,y)

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Display Plot
plt.show()
```

## **Add Legends**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot With Legend
plt.plot(x,y,label = '2x')  # Label Values Added

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Change Plot Line Properties**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot With Legend
plt.plot(x,y,label = '2x',
        color='red',linewidth=3, linestyle = '--',  # Change Line Properties
        marker = 'o',markersize=10,markeredgecolor = 'y',  # Add Markers
        )

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Change Line Style Using Shorthand Notation**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot With Legend
# Shorthand Notation '[color][marker][linestyle]'
plt.plot(x,y,'r.--',label = '2x',)

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Add Second Line To Plot**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot With Legend
# Shorthand Notation '[color][marker][linestyle]'
plt.plot(x,y,'b.--',label = '2x',)

# Add Line No. 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,color='r',label='x^2')

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Plot The Spliced Vector To Indicate Actual Vs Forecasted Line**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Create Plot With Legend
# Shorthand Notation '[color][marker][linestyle]'
plt.plot(x,y,'b.--',label = '2x',)

# Add Line No. 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2 Actual')      # Actual Data Series With Solid Line
plt.plot(x2[6:],x2[6:]**2,'r--',label='x^2 Forecasted')     # Forecasted Data Series With Dashed Line

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Change Plot Size & DPI**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Plot Size & DPI
plt.figure(figsize=(5,3),dpi=100)

# Create Plot With Legend
# Shorthand Notation '[color][marker][linestyle]'
plt.plot(x,y,'b.--',label = '2x',)

# Add Line No. 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2 Actual')      # Actual Data Series With Solid Line
plt.plot(x2[6:],x2[6:]**2,'r--',label='x^2 Forecasted')     # Forecasted Data Series With Dashed Line

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Display Plot
plt.show()
```

## **Save & Export Plot**

```python
# Define 1D Vectors
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Plot Size & DPI
plt.figure(figsize=(5,3),dpi=100)

# Create Plot With Legend
# Shorthand Notation '[color][marker][linestyle]'
plt.plot(x,y,'b.--',label = '2x',)

# Add Line No. 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2 Actual')      # Actual Data Series With Solid Line
plt.plot(x2[6:],x2[6:]**2,'r--',label='x^2 Forecasted')     # Forecasted Data Series With Dashed Line

# Add Plot & Axes Lables
plt.title('Out First Plot!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Vector X')
plt.ylabel('Vector Y')

# Add Axes Ticks
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# Add Legend
plt.legend()    # Legend Display Enabled

# Save & Export Plot
plt.savefig('Line Plot.png',dpi=300)

# Display Plot
plt.show()
```

## **Use For Loop**

* To Plot Prices For All The Countries
* Include Additional Values On X Axis Ticks

```python
# Plot Size
plt.figure(figsize=(10,7),dpi=75)

# Plot Gas Prices
for Country in gas.columns:
    if Country != 'Year':
        plt.plot(gas['Year'],gas[Country])

# Title & Labels
plt.title('Global Gas Prices')
plt.xlabel('Years')
plt.ylabel('Price (USD)')
plt.legend(loc='best')

plt.xticks([1987]+gas['Year'][::3].to_list()+[2011])

plt.show()
```

# **Bar Plot**

```python
x = ['A','B','C']  # Category Labels
y = [1,4,2]        # Category Values
```

```python

# Set Plot Size & DPI
plt.figure(figsize=(5,3),dpi=100,facecolor='grey',edgecolor='red',frameon=False,clear=False)

# Create Bar Plot
plot_02 = plt.bar(x,y)

# Add Hatching
plot_02[0].set_hatch('/')
plot_02[1].set_hatch('o')
plot_02[2].set_hatch('*')

# Display Plot
plt.show()
```

## **Use For Loop To Add Hatching**

```python
# Set Plot Size & DPI
plt.figure(figsize=(5,3),dpi=100,facecolor='grey',edgecolor='red',frameon=False,clear=False)

# Create Bar Plot
plot_02 = plt.bar(x,y)

# Add Hatching
Hatching_Patterns = ['/','o','*']
for bar in plot_02:
    bar.set_hatch(Hatching_Patterns.pop(0))

# Display Plot
plt.show()
```
