```python
# Add Total_SAT Column
schools_copy = schools.copy()
schools_copy['total_SAT'] = schools_copy['average_math'] + schools_copy['average_reading'] + schools_copy['average_writing']

print(schools_copy.shape)
schools_copy.head()
```

```python
# Group By Boroughs
largest_std_dev = schools_copy.groupby(by='borough').aggregate({'school_name':['count'],'total_SAT':['mean','std']})

# Assign Colum Names
largest_std_dev.columns = ['num_schools','average_SAT','std_SAT']

# Round Values
largest_std_dev['average_SAT'] = round(largest_std_dev['average_SAT'],2)
largest_std_dev['std_SAT'] = round(largest_std_dev['std_SAT'],2)

# Sort By Std_SAT
largest_std_dev.sort_values('std_SAT',ascending=False,inplace=True)
largest_std_dev
```
