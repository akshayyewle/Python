# **Setup**

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

# **Data Import**

```python
# Load Data
netflix_df = pd.read_csv("netflix_data.csv")

print(netflix_df.shape)
netflix_df.head()
```

# **Filter Out TV Shows**

```python
netflix_subset = netflix_df[netflix_df['type']!='TV Show']
print(netflix_subset.shape)
netflix_subset.head()
```

# **Create Netflix Movies Dataframe**

```python
netflix_movies = netflix_subset.loc[:,['title','country','genre','release_year',
                                       'duration']]
print(netflix_movies.shape)
netflix_movies.head()
```

# **Short Movies**

```python
short_movies = netflix_movies[netflix_movies['duration']<60]

print(short_movies.shape)
short_movies.head()
```

```python
short_movies['genre'].unique()
```

# **Assign Color**

```python
# Initialize Empty List
colors = list()

# Assign Colors
for i in netflix_movies.index:
    Movie_Genre = netflix_movies.at[i,'genre']
    # print(Movie_Genre)
    
    if Movie_Genre == 'Children':
        colors.append('Red')
    elif Movie_Genre == 'Documentaries':
        colors.append('Green')
    elif Movie_Genre == 'Stand-Up':
        colors.append('Blue')
    else:
        colors.append('Grey')

print(len(colors))
print(colors)
```

# **Scatter Plot**

```python
print(netflix_movies.shape)
print(netflix_movies.columns)

netflix_movies.head()
```

```python
fig = plt.figure()

plt.scatter(x=netflix_movies['release_year'],y=netflix_movies['duration'],c=colors)
plt.xlabel('Movie Duration By Release Year')
plt.ylabel('Release Year')
plt.title('Duration (Min)')
plt.show()
```
