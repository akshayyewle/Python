# **GroupBy Example**

```python
df.groupby('Continent').sum(numeric_only=True).sort_values('2022 Population',ascending=False)
```

```python
# Which Continent Has Grown The Fastest
df_02 = df.groupby('Continent').sum(numeric_only=True).sort_values('2022 Population',ascending=False)
df_02
```

```python
df_02.plot(figsize=(8,5))
```

```python
# We Need Continent Wise Plot
# Transpose The Data
df_03 = df_02.transpose()
df_03
```

```python
df.columns
```

```python
# Recreate df_02 To Select Only Population Data Columns
df_02 = df.groupby('Continent')[df.columns[5:13]].sum(numeric_only=True).sort_values('2022 Population',ascending=False)
df_02
```

```python
df_02_Selected_Columns = ['1970 Population','1980 Population','1990 Population',
                '2000 Population','2010 Population','2015 Population','2020 Population',
                '2022 Population']
df_02_Selected_Columns
```

```python
# Recreate df_02 To Select Only Population Data Columns
df_02 = df.groupby('Continent')[df_02_Selected_Columns].sum(numeric_only=True).sort_values('2022 Population',ascending=False)
df_02
```

```python
df_03 = df_02.transpose()
df_03
```

```python
# Compare Continent Wise Population Growth
df_03.plot(figsize=(12,5))
```
