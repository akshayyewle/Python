# **Data Import**

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df.head()
```

```python
df = df.set_index('Date')
df.head()
```

# **Line Plot**

```python
df.plot()   # Creates Line Plot By Default
```

```python
df.plot(kind='line',figsize=(6,4))
```

```python
# Create Sub Plots
df.plot(kind='line',subplots=True)
```

```python
# Plot Detailing
df.plot(kind='line',title='Ice Cream Ratings',xlabel='Daily Ratings',ylabel='Scores')
```

# **Bar Plot**

```python
df.plot(kind='bar')
```

## **Stacked Bar Chart**

```python
# Enable Bar Stacking
df.plot(kind='bar',stacked=True)
```

## **Horizontal Bar Chart**

```python
df.plot.barh(stacked=True)
```

# **Scatter Plot**

```python
df.plot.scatter(x='Flavor Rating',y='Texture Rating',
            s=100, #Marker Size
            c = 'Red' # Marker Color
            )
```

# **Histogram Plot**

```python
df.plot.hist(bins=20 # No. Of Bins
            )
```

# **Box Plot**

```python
df.boxplot()
```

# **Area Plot**

```python
df.plot.area()
```

# **Pie Chart**

```python
df.plot.pie(y='Flavor Rating',figsize=(6,4))
```
